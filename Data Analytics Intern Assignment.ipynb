### EDA

Average ratings for completed orders

```python
orders_completed = final_data[final_data['Order Status']=='Completed']

avg_ratings = orders_completed.groupby(['User_ID','Session ID'])['Session Rating'].mean()
avg_ratings
```

Most ordered Dishes

```python
popular_dishes = final_data.groupby('Dish_name').agg(Total_orders=('Order ID','count'),
                                                        avg_ratings=('Rating','mean')).sort_values(by='Total_orders',ascending=False)
```

```python
popular_dishes
```

Order Trend by location

```python
order_location = final_data.groupby('Location').agg(Total_orders=('Order ID','count'),
                                                    avg_order_amount=('Amount (USD)','mean')).sort_values(by='Total_orders',ascending=False)
```

Count of Favorite meals

```python
favorite_meal_orders = final_data.groupby('Favorite Meal')['Order ID'].count().reset_index()
```

```python
favorite_meal_orders
```

```python
final_data['Is_Favorite_Meal'] = final_data['Favorite Meal'] == final_data['Meal_Type']

favorite_meal_percentage = final_data['Is_Favorite_Meal'].mean() * 100

print(f"Percentage of orders matching the favorite meal: {favorite_meal_percentage:.2f}%")
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Favorite Meal', y='Order ID', data=favorite_meal_orders, palette='Oranges_d')
plt.title("Total Orders by Favorite Meal")
plt.xlabel("Favorite Meal")
plt.ylabel("Number of Orders")
plt.show()
```

Session Ratings by Age Group

```python
age_group_ratings = final_data.groupby('Age')['Session Rating'].mean().reset_index()
```

```python
age_group_ratings
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Age', y='Session Rating', data=age_group_ratings, palette='Purples_d')
plt.title("Average Session Ratings by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Session Rating")
plt.show()
```
